## Project Title

### Student Performance Analysis Using NumPy

### 1.Problem Statement

- Educational institutions often need to analyze student performance across different subjects to understand overall academic trends. Manual analysis of large datasets can be time-consuming and inefficient.

- Therefore, there is a need for a computational approach to analyze student marks efficiently and extract meaningful insights.

- In this project, we use NumPy, a powerful numerical computing library in Python, to analyze student marks and identify patterns in academic performance.

### 2.Objective

The main objectives of this project are:

- To understand how NumPy works with numerical datasets

- To perform statistical analysis on student marks

- To compute mean, median, and standard deviation

- To identify highest and lowest scores

- To analyze subject-wise performance

- To gain insights about student academic trends


### 3.Dataset Description

- The dataset contains information about 200 students and their marks in different subjects.

- All values are numerical, which makes this dataset suitable for NumPy numerical analysis.


### 4.Import Required Libraries

```python
import numpy as np
import pandas as pd
```

Explanation:

- NumPy → used for numerical computations

- Pandas → used to load the dataset

### 5.Load the Dataset

```python
df = pd.read_csv("student_marks.csv")

df.head()
```

### 6.Convert Data into NumPy Array

- NumPy works with arrays, so we convert the dataset.

```python
data = df.to_numpy()

data
```

### 7.Attributes of Numpy (Basic Functions)

```python
data.shape # retrun no rows and columns
```

```python
data.ndim # return array dimension
```

```python
data.size # return total element present in array
```

```python
data.dtype # return which type of ele present in array
```

### 8.Extract Subject Marks

- We only need subject marks (ignore Student_ID).

```python
marks = data[:,1:6]

marks
```

### 9.Calculate Average Marks of Each Subject

```python
average_marks = np.mean(marks, axis=0)

average_marks 
```

Explanation:

- axis = 0 → column wise calculation

- This gives the average marks of each subject.

### 10. Find Highest Marks in Each Subject

```python
highest_marks = np.max(marks, axis=0)

highest_marks
```

- This shows the top score in every subject.

### 11. Find Lowest Marks

```python
lowest_marks = np.min(marks, axis=0)

lowest_marks
```

- This identifies minimum marks in each subject.

### 12. Calculate Standard Deviation

```python
std_dev = np.std(marks, axis=0)

std_dev
```

- Standard deviation helps understand score variation among students.

### 13. Find Overall Average of Each Student

```python
student_average = np.mean(marks, axis=1)

student_average[:10]
```

Axis explanation:

- axis = 1 → row wise calculation

### 14. Identify Top Performing Students

```python
top_students = np.where(student_average > 85)

top_students
```

### 15. Identify Students Needing Improvement

```python
low_students = np.where(student_average < 40)

low_students
```

### 16. Total Marks of Each Student

- This calculates the sum of marks across all subjects for each student.

```python
total_marks = np.sum(marks, axis=1)

total_marks[:10]
```

Explanation:

- axis=1 → row-wise calculation

- Gives total marks of each student.

- Use case: helps identify overall academic performance.

### 17. Overall Class Average

- Calculate the average marks of the entire class.

```python
class_average = np.mean(marks)

class_average
```

- This gives the overall mean score of all students across all subjects.

### 18. Identify Topper of the Class

- Find the student with highest total marks.

```python
topper_index = np.argmax(total_marks)

topper_index

```

Explanation:

- argmax() returns the index of maximum value.

- You can also display the topper:

```python
df.iloc[topper_index]
```

### 19.Identify Weakest Student

- Find student with lowest total marks.

```python
weak_student = np.argmin(total_marks)

df.iloc[weak_student]
```

### 20.Students Scoring Above Class Average

- Find students performing better than the class average.



```python
above_average_students = np.where(student_average > class_average)

above_average_students
```

Explanation:

- Returns indices of students performing above average.

### 21. Subject-wise Highest Score Index

- Find which student scored highest in each subject. 

```python
subject_toppers = np.argmax(marks, axis=0)

subject_toppers
```

Explanation:

- Each value represents student index who scored highest in that subject.

### 22. Sort Students by Average Marks

- Sort students by academic performance.

```python
sorted_students = np.sort(student_average)

sorted_students[:10]
```

### 23. Count Students Scoring Above 80

- Find number of high-performing students.


```python
high_performers = np.sum(student_average > 80)

high_performers
```

### 24. Insights

From the analysis, we can observe:

- Some subjects have higher average marks, indicating better student performance.

- Standard deviation shows variation in student scores.

- A group of students consistently scores above 85, indicating high academic performance.

- The class topper achieved the highest overall marks across all subjects.

- A number of students are performing above the class average, indicating good academic performance.

- A small group of students requires academic improvement and support.

### 25. Recommendations

Based on the analysis:

- Teachers should provide extra support to low-performing students.

- Subjects with lower average marks may require improved teaching strategies.

- High-performing students could be encouraged with advanced learning opportunities.

- Attendance should be monitored since it may impact academic performance.
